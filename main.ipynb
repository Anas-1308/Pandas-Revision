### Filtering in Pandas

```python
df[df['Category']=="Electronics"]
```

```python
df[(df['Category']=='Electronics') & (df['Country']=='Japan')]
```

```python
df[(df['Category']=="Electronics") | (df['Country']=="India")]
```

```python
df[(df['Price']<=100.0) & (df['Shipped']=='No')]
```

```python
# df[df['CustomerName'].str.startswith("A")]
df[df['CustomerName'].str.endswith("a")]
```

```python
df[df['Country'].isin(["UAE", "India", "Brazil"])]
```

```python
df[~df['Country'].isin(["UAE", "India", "Brazil"])]
```

### Data Modification: Updation
.loc():- means locating rows via labels like name product etc rather than the index


```python
df.loc[df['CustomerName']=='Isabella Silva',"Shipped"]="No"
```

```python
df.loc[df["CustomerName"]=="Isabella Silva"]
```

```python
df.loc[df["Country"]=='USA',"Country"]="Uninted States"
```

```python
df.loc[df['Country']=='Uninted States']
```
